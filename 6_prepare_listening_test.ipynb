# Notebook for preparing files for listening test

```python
# libs
import numpy as np
import matplotlib.pyplot as plt
import json
import os
import glob
import librosa
import soundfile as sf
```

```python
# path to wav files
paths = {
    'produced': '/Users/marius/Documents/Uni/TU_Berlin_Master/Masterarbeit/Data/listeningTestFiles/0_Produced/*.wav',
    'noisy': '/Users/marius/Documents/Uni/TU_Berlin_Master/Masterarbeit/Data/listeningTestFiles/1_noisy/*.wav',
    'voicefixer': '/Users/marius/Documents/Uni/TU_Berlin_Master/Masterarbeit/Data/listeningTestFiles/2_Voicefixer/*.wav',
}
```

```python
# load voicefixer files and get n samples with duration 5s from it
n = 5

# cut files
for file in glob.glob(paths['voicefixer']):
    for k in range(n):
        # load file
        y, sr = librosa.load(file, sr=44100, mono=True, duration=5.0, offset=(k+1)*6.0)
        # save file
        # get save path
        save_path = os.path.join(os.path.dirname(file), os.path.basename(file).split('.')[0] + '_5s_n' + str(k+1) + '.wav')
        sf.write(save_path , y, sr, subtype='PCM_16')
```

```python
# load produced files and get n samples with duration 5s from it
n = 5

# cut files
for file in glob.glob(paths['produced']):
    for k in range(n):
        # load file
        y, sr = librosa.load(file, sr=44100, mono=True, duration=5.0, offset=(k+1)*6.0)
        # save file
        # get save path
        save_path = os.path.join(os.path.dirname(file), os.path.basename(file).split('.')[0] + '_5s_n' + str(k+1) + '.wav')
        sf.write(save_path , y, sr, subtype='PCM_16')
```

```python
# load noisy files and get n samples with duration 5s from it
n = 5

# cut files
for file in glob.glob(paths['noisy']):
    for k in range(n):
        # load file
        y, sr = librosa.load(file, sr=44100, mono=True, duration=5.0, offset=(k+1)*6.0)
        # save file
        # get save path
        save_path = os.path.join(os.path.dirname(file), os.path.basename(file).split('.')[0] + '_5s_n' + str(k+1) + '.wav')
        sf.write(save_path , y, sr, subtype='PCM_16')
```

# Load and predict with models

```python
import tensorflow as tf
from tensorflow import keras

# path to wav files
paths = {
    'voicefixer': '/Users/marius/Documents/Uni/TU_Berlin_Master/Masterarbeit/Data/listeningTestFiles/2_Voicefixer/*.wav',
}

path_inputs = glob.glob(paths['voicefixer'])

mix2_folder = '/Users/marius/Documents/Uni/TU_Berlin_Master/Masterarbeit/Data/listeningTestFiles/4_PP_Net_1/'
mix4_folder = '/Users/marius/Documents/Uni/TU_Berlin_Master/Masterarbeit/Data/listeningTestFiles/5_PP_Net_2/'

model2 = keras.models.load_model(mix2_folder + 'mix2.keras', compile=False)
model4 = keras.models.load_model(mix4_folder + 'mix4.keras', compile=False)

# get model output and save to folders
for file in path_inputs:
    print(file)

    # load file
    y, sr = librosa.load(file, sr=22050, mono=True)

    # add dimension
    y = np.expand_dims(y, axis=-1)
    print(y.shape)

    # predict 1
    output2 = model2.predict(y)
    # get file and delete 'voicefixer'
    f = os.path.basename(file).split('_')
    del f[-3]
    save_path = mix2_folder + '_'.join(f)
    # save file
    sf.write(save_path, np.squeeze(output2), sr, subtype='PCM_16')

    # predict 2
    output4 = model4.predict(y)
    # get file and delete 'voicefixer'
    f = os.path.basename(file).split('_')
    del f[-3]
    save_path = mix4_folder + '_'.join(f)
    # save file
    sf.write(save_path , np.squeeze(output4), sr, subtype='PCM_16')


```
